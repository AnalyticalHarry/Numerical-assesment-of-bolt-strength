### 1. Mechanical Bolt Forces

When a bolt is tightened, multiple different kinds of mechanical force are activated, aiding in the stability of the resulting joint.

1. Clamping force: The bolt delivers a form of compression to the joint to hold the two parts together.
2. Preloading Force: Turning the bolt causes the threads to engage and stretch, which produces the preload force that keeps the threads in place.
3. Shear Force: The shear force is the transverse pressure working against the bolt in a perpendicular direction. It can be present alone or in combination with tension force.
4. Tension Force: Tension force applies to the length of the bolt, providing pressure along its vertical dimension, rather than its width.

The clamping force accomplishes the task of holding multiple parts together, and is determined by the difference between the preloading force and the tension force. When the strength of the tension force equals that of the preload force, the bolt fails and the joint breaks apart. Generally, higher preload force translates into higher clamping force, which makes for a stronger bolted joint. Ideally, precise bolt installation and tightening will create a joint that yields little or no direct shear force.

### 2. Tensile Strength of Bolt

```python
force = float(input("Enter force(Newton): "))
diameter_of_bolt = float(input("Enter diameter of bolt(mm): "))
material_yield_strength = float(input("Enter yield strength(Mpa or N/mm2): "))
material_shear_strength = float(input("Enter sheer strength(Mpa or N/mm2): "))

pie = 22/7
area = (pie/4)*diameter_of_bolt*diameter_of_bolt
tensile_stress = force/area

factor_of_safety = material_yield_strength/tensile_stress
```

```python
print("Input:\n")
print(f"Tensile load: {force}N")
print(f"Diameter of bolt: {diameter_of_bolt}mm")
print(f"Material Yield Strength: {material_yield_strength}N/mm2")
print(f"Material Shear Strength: {material_shear_strength}N/mm2")
print("\nOutput:")
print(f"\nArea calulated: {round(area, 4)}mm2")
print(f"Tensile Stress calculated: {round(tensile_stress, 4)}N/mm2")
print(f"Factor of safety calculated: {round(factor_of_safety, 4)} ")
```

### 3. Single shear strength of bolt

In single shear, load is applied on only on one plane.

```python
force = float(input("Enter force(Newton): "))
diameter_of_bolt = float(input("Enter diameter of bolt(mm): "))
material_yield_strength = float(input("Enter yield strength(Mpa or N/mm2): "))
material_shear_strength = float(input("Enter sheer strength(Mpa or N/mm2): "))

pie = 22/7
area = (pie/4)*diameter_of_bolt*diameter_of_bolt
shear_stress = force/area

factor_of_safety = material_shear_strength/shear_stress
```

```python
print("Input:\n")
print(f"Tensile load: {force}N")
print(f"Diameter of bolt: {diameter_of_bolt}mm")
print(f"Material Yield Strength: {material_yield_strength}N/mm2")
print(f"Material Shear Strength: {material_shear_strength}N/mm2")
print("\nOutput:")
print(f"\nArea calulated: {round(area, 4)}mm2")
print(f"Shear Stress calculated: {round(shear_stress, 4)}N/mm2")
print(f"Factor of safety calculated: {round(factor_of_safety, 4)} ")
```

### 4. Double Shear Strength of Bolt

In double shear, load is applied on two planes.

```python
force = float(input("Enter force(Newton): "))
diameter_of_bolt = float(input("Enter diameter of bolt(mm): "))
material_yield_strength = float(input("Enter yield strength(Mpa or N/mm2): "))
material_shear_strength = float(input("Enter sheer strength(Mpa or N/mm2): "))

pie = 22/7
area = (pie/2)*diameter_of_bolt*diameter_of_bolt
double_shear = force/area

factor_of_safety = material_shear_strength/double_shear
```

```python
print("Input:\n")
print(f"Tensile load: {force}N")
print(f"Diameter of bolt: {diameter_of_bolt}mm")
print(f"Material Yield Strength: {material_yield_strength}N/mm2")
print(f"Material Shear Strength: {material_shear_strength}N/mm2")
print("\nOutput:")
print(f"\nArea calulated: {round(area, 4)}mm2")
print(f"Double shear force calculated: {round(double_shear, 4)}N/mm2")
print(f"Factor of safety calculated: {round(factor_of_safety, 4)} ")
```
